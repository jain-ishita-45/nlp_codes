```python
from nltk.stem import PorterStemmer
from nltk import sent_tokenize
stemmer=PorterStemmer()
para="A good paragraph in English, such as this example from Study.com, contains a topic sentence, supporting details, and a conclusion. It is a series of sentences that are organized, coherent, and relate to a single topic. For example, There are many different kinds of animals that live in China. Tigers and leopards are animals that live in China's forests in the north. In the jungles, monkeys swing in the trees and elephants walk through the brush"
text=nltk.sent_tokenize(para)
print(type(text))
```

```python
#apply stopwords and filter and then apply stemming

stop_words = set(stopwords.words('english'))

for i in range(len(text)):
    words = nltk.word_tokenize(text[i])
    words = [
        stemmer.stem(word)
        for word in words
        if word.lower() not in stop_words and word.isalpha()
    ]
    text[i] = ' '.join(words)
print(text)
```

```python
#find the pos tags
for i in range(len(text)):
    words=nltk.word_tokenize(text[i])
    words=[word for word in words if word not in stop_words]
    pos_tag=nltk.pos_tag(words)
    print(pos_tag)
```

```python
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger
from nltk.corpus import treebank
testing = treebank.tagged_sents()[2000:]
training= treebank.tagged_sents()[:7000]
tag1=DefaultTagger('NN')
tag2=UnigramTagger(training,backoff=tag1)
print(tag2.evaluate(testing))
```

```python
from nltk import Nonterminal, nonterminals, Production, CFG
nonterminal1 = Nonterminal('NP')
nonterminal2 = Nonterminal('VP')
nonterminal3 = Nonterminal('PP')
print(nonterminal1.symbol())
print(nonterminal2.symbol())
print(nonterminal3.symbol())
print(nonterminal1==nonterminal2)
print(nonterminal2==nonterminal3)
print(nonterminal1==nonterminal3)
S, NP, VP, PP = nonterminals('S, NP, VP, PP')
N, V, P, DT = nonterminals('N, V, P, DT')
production1 = Production(S, [NP, VP])
production2 = Production(NP, [DT, NP])
production3 = Production(VP, [V, NP,NP,PP])
print(production1.lhs())
print(production1.rhs())
print(production3.lhs())
print(production3.rhs())
print(production3 == Production(VP, [V,NP,NP,PP]))
print(production2 == production3)
```
